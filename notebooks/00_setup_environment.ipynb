# 00. Environment Setup

Install system dependencies and the `longreadclock` Python package.

Run this notebook **once** after cloning the repository into your All of Us Workbench.
All subsequent notebooks assume these tools are available.


## 1. Clone & build wgbstools

```python
%%bash
# wgbstools is required for PAT indexing and PAT→beta conversion
if [ ! -d ~/wgbs_tools ]; then
    git clone https://github.com/nloyfer/wgbs_tools.git ~/wgbs_tools
fi
cd ~/wgbs_tools && make -j4
echo "wgbstools binary: $(ls -lh ~/wgbs_tools/wgbstools)"
```

## 2. Install bgzip (tabix package)

```python
%%bash
# bgzip is needed to compress PAT files after sorting
conda install -y -c bioconda tabix 2>/dev/null || pip install htslib 2>/dev/null || true
which bgzip && bgzip --version | head -1
```

## 3. Install the longreadclock package

```python
%%bash
# Install in editable mode so notebook changes are reflected immediately
pip install -e /home/jupyter/LongReadClock_Claude --quiet
python -c "import longreadclock; print('longreadclock', longreadclock.__version__, 'installed OK')"
```
